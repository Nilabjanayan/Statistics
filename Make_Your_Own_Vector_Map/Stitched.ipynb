```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
%matplotlib inline
```

```python
p1 = cv2.imread("satellite_images/p01.png")
p2 = cv2.imread("satellite_images/p02.png")
p3 = cv2.imread("satellite_images/p03.png")
p4 = cv2.imread("satellite_images/p04.png")
p5 = cv2.imread("satellite_images/p05.png")
p6 = cv2.imread("satellite_images/p06.png")
p7 = cv2.imread("satellite_images/p07.png")
p8 = cv2.imread("satellite_images/p08.png")
p9 = cv2.imread("satellite_images/p09.png")
p10 = cv2.imread("satellite_images/p10.png")
```

```python
(status,stitched)=cv2.Stitcher_create().stitch([p1,p2,p3,p4,p5,p6,p7,p8,p9,p10])
print(status)
```

```python
plt.figure(figsize = (20,20))
if status ==0:
#    cv2.imwrite('stitched_images/stitched.png',stitched)
    plt.imshow(stitched[:,:,::-1])
else:
    print("[INFO] image stitching failed ({})".format(status))
```
